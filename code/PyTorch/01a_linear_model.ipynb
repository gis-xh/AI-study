## 1 添加必要的包

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2 添加数据

```python
# 训练数据集
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

## 3 定义模型及其损失函数

Linear Model：$\hat{y}=x*\omega$

```python
# 线性模型
global w
def linear_model(x):
    return x * w
# 损失函数
def loss(x, y):
    y_pred = linear_model(x)
    return (y_pred - y) * (y_pred - y)
```

## 4 调用模型

```python
# 记录权重和损失值的列表
w_list = []
mse_list = []
# 设置权重从 0.0 到 4.0 进行间隔为 0.1 的采样
# 0.0, 0.1, 0.2, ...
for w in np.arange(0.0, 4.1, 0.1):
    print('w =', w)
    loss_sum = 0
    # 用 zip() 函数拼接数据集列表中的数据
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = linear_model(x_val)
        loss_val = loss(x_val, y_val)
        loss_sum += loss_val
        print('\t', x_val, y_val, y_pred_val, loss_val)
    linear_mse = loss_sum / 3
    print('MSE =', linear_mse)
    # 把权重和 MSE 添加到列表中
    w_list.append(w)
    mse_list.append(linear_mse)
```

## 5 绘制损失曲线

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
