# Linear Model：$\hat{y}=x*\omega+b$

```python
# 1. 导入必要的包
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# 2. 训练数据集
x_data = [1.0, 2.0, 3.0]
y_data = [1.9, 3.9, 5.9]

# 3. 定义模型及其损失函数
# 设置模型参数范围，保留一位小数，生成两个一维坐标矩阵
W = np.arange(0.0, 4.1, 0.1).round(1)
B = np.arange(-2.0, 2.1, 0.1).round(1)
# 构建二维坐标矩阵，表示 x,y 轴坐标网格点
w, b = np.meshgrid(W, B)
# 线性模型
def linear_model(x):
    return x * w + b
# 损失函数
def loss(y_pred, y):
    return (y_pred - y) ** 2

# 4. 调用模型并计算 MSE
loss_sum = 0
# 用 zip() 函数拼接数据集列表中的数据
for x_val, y_val in zip(x_data, y_data):
    y_pred_val = linear_model(x_val)
    loss_val = loss(y_pred_val, y_val)
    loss_sum += loss_val
    # print('\t', x_val, y_val, y_pred_val, loss_val)
linear_mse = loss_sum / len(x_data)

# 5. 绘制损失函数曲线
# 定义figure
fig = plt.figure(figsize=(10,10), dpi=300)
# 绘制 3d 图
ax = plt.axes(projection='3d')
# 绘制表面图，rstride:行之间的跨度  cstride:列之间的跨度
surf = ax.plot_surface(w, b, linear_mse, rstride=1, cstride=1, cmap=cm.coolwarm)
# 设置 z 轴
ax.set_zlim(0, 39)
# 设置坐标轴标签
ax.set_xlabel("w")
ax.set_ylabel("b")
# ax.set_zlabel("loss")
ax.text(0.2, 2, 43, "Cost Value", color='black')
# 添加颜色映射条
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
