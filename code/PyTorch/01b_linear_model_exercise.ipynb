# Linear Modelï¼š$\hat{y}=x*\omega+b$

```python
# 1. å¯¼å…¥å¿…è¦çš„åŒ…
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# 2. è®­ç»ƒæ•°æ®é›†
x_data = [1.0, 2.0, 3.0]
y_data = [1.9, 3.9, 5.9]

# 3. å®šä¹‰æ¨¡åž‹åŠå…¶æŸå¤±å‡½æ•°
# è®¾ç½®æ¨¡åž‹å‚æ•°èŒƒå›´ï¼Œä¿ç•™ä¸€ä½å°æ•°ï¼Œç”Ÿæˆä¸¤ä¸ªä¸€ç»´åæ ‡çŸ©é˜µ
W = np.arange(0.0, 4.1, 0.1).round(1)
B = np.arange(-2.0, 2.1, 0.1).round(1)
# æž„å»ºäºŒç»´åæ ‡çŸ©é˜µï¼Œè¡¨ç¤º x,y è½´åæ ‡ç½‘æ ¼ç‚¹
w, b = np.meshgrid(W, B)
# çº¿æ€§æ¨¡åž‹
def linear_model(x):
    return x * w + b
# æŸå¤±å‡½æ•°
def loss(y_pred, y):
    return (y_pred - y) ** 2

# 4. è°ƒç”¨æ¨¡åž‹å¹¶è®¡ç®— MSE
loss_sum = 0
# ç”¨ zip() å‡½æ•°æ‹¼æŽ¥æ•°æ®é›†åˆ—è¡¨ä¸­çš„æ•°æ®
for x_val, y_val in zip(x_data, y_data):
    y_pred_val = linear_model(x_val)
    loss_val = loss(y_pred_val, y_val)
    loss_sum += loss_val
    # print('\t', x_val, y_val, y_pred_val, loss_val)
linear_mse = loss_sum / len(x_data)

# 5. ç»˜åˆ¶æŸå¤±å‡½æ•°æ›²çº¿
# å®šä¹‰figure
fig = plt.figure(figsize=(10,10), dpi=300)
# ç»˜åˆ¶ 3d å›¾
ax = plt.axes(projection='3d')
# ç»˜åˆ¶è¡¨é¢å›¾ï¼Œrstride:è¡Œä¹‹é—´çš„è·¨åº¦  cstride:åˆ—ä¹‹é—´çš„è·¨åº¦
surf = ax.plot_surface(w, b, linear_mse, rstride=1, cstride=1, cmap=cm.coolwarm)
# è®¾ç½® z è½´
ax.set_zlim(0, 39)
# è®¾ç½®åæ ‡è½´æ ‡ç­¾
ax.set_xlabel("w")
ax.set_ylabel("b")
# ax.set_zlabel("loss")
ax.text(0.2, 2, 43, "Cost Value", color='black')
# æ·»åŠ é¢œè‰²æ˜ å°„æ¡
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
