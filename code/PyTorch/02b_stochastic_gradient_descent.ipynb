```python
import numpy as np
import matplotlib.pyplot as plt

# 准备训练集
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
# 初始化权重
w = 1.0
# 定义线性模型
def linear_Model(x):
    return x * w
```

## 1 定义 loss 函数

$$
loss(\omega)=(\hat{y}-y)^2=(x*\omega-y)^2
$$

```python
def loss(x, y):
    y_pred = linear_Model(x)
    loss_val = (y_pred - y) ** 2
    return loss_val
```

## 2 随机梯度函数

$$
\frac{\partial {loss}_i}{\partial \ \omega}
=2\cdot{x_i}\cdot({x_i}\cdot\omega-{y_i})
$$

```python
def gradient(x, y):
    grad = 2 * x * (x * w - y)
    return grad
```

## 3 随机梯度下降

$$
\omega=\omega-\alpha\frac{\partial{loss}}{\partial{\omega}}
$$

```python
a = 0.195
print('Predict (before training)', 4, linear_Model(4))
epoch_list = []
loss_list = []
for epoch in range(100):
    for x, y in zip(x_data, y_data):
        grad_val = gradient(x, y)
        w -= a * grad_val
        l = loss(x, y)
        print('\tgrad:', x, y, grad_val)
        epoch_list.append(epoch)
        loss_list.append(l)

print('Predict (after training)', 4, linear_Model(4))
```

## 6 绘制图像

```python
plt.plot(epoch_list, loss_list)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.grid()
plt.show()
```
